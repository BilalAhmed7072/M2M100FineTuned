```python
from transformers import AutoModelForSeq2SeqLM , AutoTokenizer
```

```python
repo_id = "Bilal7072/m2m100-finetuned-tech"
```

```python
tokenizer = AutoTokenizer.from_pretrained(repo_id)
model = AutoModelForSeq2SeqLM.from_pretrained(repo_id)
```

```python
text = "As a trusted web and app development company, we specialize in delivering innovative, high-quality solutions tailored to your business needs. At Spark Solutionz, we prioritize client objectives, ensuring precision, creativity, and excellence in every project."
```

```python
tokenizer.src_lang = "en"
```

```python
inputs = tokenizer(text, return_tensors="pt")
```

```python
generated= model.generate(**inputs, max_length=128, forced_bos_token_id=tokenizer.get_lang_id("ur"))
```

```python
translation = tokenizer.decode(generated[0], skip_special_tokens=True)
print(translation)
```
